```python
import numpy as np
import pandas as pd
import plotly.plotly as py
import plotly.graph_objs as go
from plotly.offline import init_notebook_mode,iplot
import cufflinks as cf
import IPython
import plotly.figure_factory as ff
init_notebook_mode(connected=True)
cf.go_offline(connected=True)
cf.set_config_file(offline=False, world_readable=True, theme='ggplot')
def cell():
    display(IPython.core.display.HTML('''<script src="/static/components/requirejs/require.js"></script>'''))
```

```python
from sklearn.datasets import load_iris
```

```python
df=load_iris()
```

```python
target_names=df.target_names.tolist()
feature_names=df.feature_names
target=df.target
data=df.data
df_d=pd.DataFrame(data=data,columns=feature_names)
df_d['class']=list(map(str,target))
```

```python
df_d.sample(10)
```

```python
df_d['class'].iplot(kind='histogram',
                          histnorm='',
                          barmode='',
                          theme='solar',
                          bins=5,
                          xTitle='Class Type',
                          yTitle='Count')
```

```python
df_d[feature_names].iplot(kind='hist',
                          subplots=True,
                          shape=(4,1),
                          histnorm='probability density',
                          #histfunc='count',
                          theme='ggplot')
```

```python
df_d1=df_d[['sepal length (cm)','sepal width (cm)','class']].set_index('class')
df_d1.iplot(kind='barh',
            barmode='stack',
            bargap=.01,
            theme='pearl')
```

```python
df_d[feature_names].iplot(theme='space',bestfit=True,bestfit_colors=['azure','antiquewhite','orchid','silver'])
```

```python
df_d.iplot(kind='scatter',mode='markers+text',x='sepal length (cm)',
           y='petal length (cm)',
           categories='class',
           xTitle='sepal length',
           yTitle='Petal Length',
           colors=['ivory','dodgerblue','lawngreen'],
           theme='solar')
```

```python
df_d.iplot(kind='scatter',mode='markers',x='sepal length (cm)',
           y='petal length (cm)',
           size=[int(x)*5+5 for x in df_d['class']],
           xTitle='sepal length',
           yTitle='Petal Length',
           theme='ggplot')
```

```python
df_d[feature_names].iplot(kind='box')
```

```python
df_d[feature_names].iplot(kind='area',fill=True,theme='space')
```

```python
corr=df_d[feature_names].corr()
```

```python
figure = ff.create_annotated_heatmap(
    z=corr.values,
    x=list(corr.columns),
    y=list(corr.index),
    annotation_text=corr.round(2).values,
    colorscale='Viridis',
    showscale=True)
iplot(figure)
```

```python
colorscale=[[0,'rgb(55,35,23)'],[0.3,'rgb(164,56,32)'],[0.7,'rgb(245,43,65)'],[1,'rgb(255,25,24)']]
figure = ff.create_annotated_heatmap(
    z=abs(corr.values),
    x=list(corr.columns),
    y=list(corr.index),
    annotation_text=abs(corr.round(2).values),
    colorscale=colorscale,
    showscale=True)
iplot(figure)
```

```python
figure = ff.create_scatterplotmatrix(
    df_d,
    diag='box',#(scatter,histogram,box)
    index='class',
    colormap='Portland',
    colormap_type='seq',
    height=700,
    width=800)
iplot(figure)
```

```python
figure = ff.create_table(df_d.sample(10),colorscale='Blackbody')
iplot(figure)
```
